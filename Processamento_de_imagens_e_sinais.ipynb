- Imagem 1

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

imagem=cv2.imread("/content/drive/MyDrive/Imagens/MulherRuiva.png")

imagem_sem_ruido = cv2.medianBlur(imagem, 7)

imagem_filtrada = cv2.GaussianBlur(imagem_sem_ruido, (5, 5), 0)

# Plotar a imagem filtrada
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(imagem_filtrada, cv2.COLOR_BGR2RGB))
plt.title("Imagem Melhorada")
plt.axis("off")
plt.show()
```

- Imagem 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagem=cv2.imread("/content/drive/MyDrive/Imagens/Mulher.jpg")

imagem_normalizada = cv2.normalize(imagem, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

imagem_sem_ruido = cv2.medianBlur(imagem_normalizada, 3)


# Plotar a imagem sem ruído
plt.figure(figsize=(8, 6))
plt.imshow(imagem_sem_ruido, cmap="gray")
plt.title("Imagem melhorada")
plt.axis("off")
plt.show()
```
