```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("train.csv")
df.head(8)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

A quantidade de nulos em 'Age' não é irrisória, mas também não afeta muita coisa, 'Cabin' é praticamente toda nula, tendo em vista que 687/891 são nulos e a quantidade de nulos em 'Embarked' é irrisória.

```python
# Os nulos de 'Age' serão substituídos pela media adquirida com a função .describe
df=df.fillna({"Age":df["Age"].mean()})

# 'Cabin' será removida, pois tem pouquíssimas informações
df.drop('Cabin', axis=1, inplace=True)

# Os nulos de 'Embarked' serão substituídos pelo valor que mais aparece
df=df.fillna({"Embarked":df["Embarked"].mode()[0]})
```

```python
df['Embarked'].value_counts()
```

A maioria dos passageiros era de Southampton, com 644, sendo seguido por 168 de Cherbourg e 77 de Queenstown

```python
df['Sex'].value_counts()
```

Em sua maioria os passageiros eram homens, 577, e a minoria era das mulheres, com 314

```python
df=df.replace('male', 0)
df=df.replace('female', 1)
df=df.replace('S', 0)
df=df.replace('Q', 1)
df=df.replace('C', 2)
```

Com o gênero sendo representado por 1 para mulheres e 0 para homens e as cidades por 0, 1 e 2 os dados poderão ser levados em conta pela máquina.

```python
df.drop('PassengerId', axis=1, inplace=True)
df.drop('Ticket', axis=1, inplace=True)
df.drop('Name', axis=1, inplace=True)
df
```

As colunas 'PassengerId', 'Ticket' e 'Name' serão retiradas, pois não influenciam na sobrevivência dos passageiros

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Survived', axis=1)
y=df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

clf=RandomForestClassifier()
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```
